Combina múltiples archivos csv que contienen datos históricos de criptomonedas para diferentes meses y años en un solo DataFrame llamado `df_total`. Además, se elimina una columna no deseada que podría haber sido creada al cargar los archivos csv. Esto facilita el análisis y la manipulación de los datos históricos de las criptomonedas.

----------

```python
df_total = df_total[df_total['fecha'] >= "2016-01"]
df_total.dropna(inplace = True)
df_total.to_csv("../data/total.csv")
df_total
```

Se aplica un filtro temporal y una limpieza de datos nulos al DataFrame df_total, se guarda la versión modificada en un archivo csv y se muestra el DataFrame resultante.
